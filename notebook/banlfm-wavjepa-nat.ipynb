````python
# ============================================================
# LFM Speech Dataset Studio v5.1 — FINAL FIXED one-cell Kaggle project
#
# Fix over v5:
#   - Fixes SameFileError in final artifact copy.
#   - Writes JSONL to results/ first, then safely copies into pr_artifacts/.
#   - Adds safe_copy() for all artifact copies.
#   - Keeps v5 design: random/clean/stress subsets, LFM-Audio capability audit,
#     BanglaASR baseline, Liquid repair potential, router abstention,
#     deterministic reviewable action cards, failure casebook, PR zip.
#
# Kaggle:
#   - Internet: ON
#   - Accelerator: T4 GPU
# ============================================================

import os, sys, json, time, math, re, gc, io, random, shutil, subprocess, warnings
from pathlib import Path
from datetime import datetime, timezone

warnings.filterwarnings("ignore")

def pip_install_missing():
    import importlib.util
    pkg_map = {
        "datasets": "datasets",
        "transformers": "transformers",
        "accelerate": "accelerate",
        "huggingface_hub": "huggingface_hub",
        "soundfile": "soundfile",
        "sklearn": "scikit-learn",
        "matplotlib": "matplotlib",
        "scipy": "scipy",
        "tabulate": "tabulate",
    }
    missing = []
    for mod, pkg in pkg_map.items():
        if importlib.util.find_spec(mod) is None:
            missing.append(pkg)
    if missing:
        print("Installing missing packages:", missing)
        subprocess.run([sys.executable, "-m", "pip", "install", "-q"] + missing, check=False)

pip_install_missing()

import numpy as np
import pandas as pd
import soundfile as sf
import matplotlib.pyplot as plt
import torch

from tqdm.auto import tqdm
from datasets import load_dataset
from huggingface_hub import hf_hub_download
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

from transformers import AutoTokenizer, AutoModelForCausalLM, AutoProcessor, AutoModelForSpeechSeq2Seq

try:
    from IPython.display import display
except Exception:
    display = None

def show(x):
    if display is not None:
        display(x)
    else:
        print(x)

# ============================================================
# 1. Config
# ============================================================
SEED = int(os.environ.get("SEED", "42"))
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

RUN_FAST = os.environ.get("RUN_FAST", "1") == "1"

RUN_LFM_AUDIO_AUDIT = os.environ.get("RUN_LFM_AUDIO_AUDIT", "1") == "1"
RUN_BANGLA_ASR = os.environ.get("RUN_BANGLA_ASR", "1") == "1"
RUN_LFM_TEXT = os.environ.get("RUN_LFM_TEXT", "1") == "1"
RUN_LFM_CARD_PROPOSALS = os.environ.get("RUN_LFM_CARD_PROPOSALS", "0") == "1"
BUILD_LFM_AUDIO_RUNNER = os.environ.get("BUILD_LFM_AUDIO_RUNNER", "1") == "1"

DATASET_ID = os.environ.get("DATASET_ID", "SUST-CSE-Speech/banspeech")
LFM_AUDIO_GGUF_REPO = os.environ.get("LFM_AUDIO_GGUF_REPO", "LiquidAI/LFM2.5-Audio-1.5B-GGUF")
LFM_AUDIO_QUANT = os.environ.get("LFM_AUDIO_QUANT", "Q4_0")
LFM_TEXT_ID = os.environ.get("LFM_TEXT_ID", "LiquidAI/LFM2.5-1.2B-Instruct")
LFM_TEXT_FALLBACK_ID = os.environ.get("LFM_TEXT_FALLBACK_ID", "LiquidAI/LFM2.5-350M")
BANGLA_ASR_ID = os.environ.get("BANGLA_ASR_ID", "bangla-speech-processing/BanglaASR")

TARGET_SR = int(os.environ.get("TARGET_SR", "16000"))

if RUN_FAST:
    RANDOM_AUDIT_N = int(os.environ.get("RANDOM_AUDIT_N", "80"))
    CLEAN_DEMO_N = int(os.environ.get("CLEAN_DEMO_N", "45"))
    STRESS_N = int(os.environ.get("STRESS_N", "40"))
    LFM_AUDIO_N = int(os.environ.get("LFM_AUDIO_N", "5"))
    LFM_REPAIR_N = int(os.environ.get("LFM_REPAIR_N", "125"))
    ACTION_CARD_N = int(os.environ.get("ACTION_CARD_N", "60"))
else:
    RANDOM_AUDIT_N = int(os.environ.get("RANDOM_AUDIT_N", "220"))
    CLEAN_DEMO_N = int(os.environ.get("CLEAN_DEMO_N", "80"))
    STRESS_N = int(os.environ.get("STRESS_N", "80"))
    LFM_AUDIO_N = int(os.environ.get("LFM_AUDIO_N", "10"))
    LFM_REPAIR_N = int(os.environ.get("LFM_REPAIR_N", "260"))
    ACTION_CARD_N = int(os.environ.get("ACTION_CARD_N", "120"))

ROUTE_CONF_THRESHOLD = float(os.environ.get("ROUTE_CONF_THRESHOLD", "0.50"))
AUTO_ROUTE_CONF_THRESHOLD = float(os.environ.get("AUTO_ROUTE_CONF_THRESHOLD", "0.65"))
MIN_ROUTE_CHARS = int(os.environ.get("MIN_ROUTE_CHARS", "12"))
MAX_AUTO_CER = float(os.environ.get("MAX_AUTO_CER", "0.35"))

LLAMA_CPP_AUDIO_PR = os.environ.get("LLAMA_CPP_AUDIO_PR", "18641")
BUILD_JOBS = int(os.environ.get("BUILD_JOBS", "2"))
LFM_AUDIO_TIMEOUT = int(os.environ.get("LFM_AUDIO_TIMEOUT", "240"))

OUT_ROOT = Path(os.environ.get("OUT_ROOT", "/kaggle/working/lfm_speech_dataset_studio_v5_1"))
DATA_DIR = OUT_ROOT / "data"
AUDIO_CACHE = OUT_ROOT / "audio_cache"
MODEL_DIR = OUT_ROOT / "models"
RESULTS_DIR = OUT_ROOT / "results"
FIGURES_DIR = OUT_ROOT / "figures"
LOG_DIR = OUT_ROOT / "logs"
PR_DIR = OUT_ROOT / "pr_artifacts"
THIRD_PARTY_DIR = OUT_ROOT / "third_party"
AUDIO_AUDIT_DIR = OUT_ROOT / "audio_audit_raw"
STDERR_DIR = LOG_DIR / "lfm_audio_raw_stderr"
STDOUT_DIR = LOG_DIR / "lfm_audio_raw_stdout"
OUTFILE_DIR = LOG_DIR / "lfm_audio_output_files"

for d in [
    OUT_ROOT, DATA_DIR, AUDIO_CACHE, MODEL_DIR, RESULTS_DIR, FIGURES_DIR,
    LOG_DIR, PR_DIR, THIRD_PARTY_DIR, AUDIO_AUDIT_DIR, STDERR_DIR, STDOUT_DIR, OUTFILE_DIR
]:
    d.mkdir(parents=True, exist_ok=True)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
DTYPE = torch.float16 if torch.cuda.is_available() else torch.float32

def now_utc():
    return datetime.now(timezone.utc).isoformat()

def to_builtin(x):
    if isinstance(x, dict):
        return {str(k): to_builtin(v) for k, v in x.items()}
    if isinstance(x, list):
        return [to_builtin(v) for v in x]
    if isinstance(x, tuple):
        return [to_builtin(v) for v in x]
    if isinstance(x, Path):
        return str(x)
    if isinstance(x, np.ndarray):
        return x.tolist()
    if isinstance(x, (np.integer,)):
        return int(x)
    if isinstance(x, (np.floating,)):
        v = float(x)
        return None if math.isnan(v) else v
    if isinstance(x, float):
        return None if math.isnan(x) else x
    return x

def safe_json_write(obj, path):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    path.write_text(json.dumps(to_builtin(obj), indent=2, ensure_ascii=False), encoding="utf-8")
    return path

def safe_copy(src, dst):
    src = Path(src)
    dst = Path(dst)
    if not src.exists():
        return None
    dst.parent.mkdir(parents=True, exist_ok=True)
    try:
        if src.resolve() == dst.resolve():
            return str(dst)
    except Exception:
        pass
    return shutil.copy2(src, dst)

def log_failure(name, error, extra=None):
    payload = {
        "created_at": now_utc(),
        "name": name,
        "error": repr(error),
        "extra": extra or {},
    }
    safe_json_write(payload, LOG_DIR / f"{name}_failure.json")
    print(f"[FAIL-LOGGED] {name}: {repr(error)}")
    return payload

def gpu_memory():
    if not torch.cuda.is_available():
        return {"cuda": False}
    try:
        torch.cuda.synchronize()
    except Exception:
        pass
    return {
        "cuda": True,
        "gpu": torch.cuda.get_device_name(0),
        "allocated_mb": float(torch.cuda.memory_allocated() / 1024**2),
        "reserved_mb": float(torch.cuda.memory_reserved() / 1024**2),
        "max_allocated_mb": float(torch.cuda.max_memory_allocated() / 1024**2),
    }

run_manifest = {
    "created_at": now_utc(),
    "project": "LFM Speech Dataset Studio v5.1",
    "dataset_id": DATASET_ID,
    "lfm_audio_gguf_repo": LFM_AUDIO_GGUF_REPO,
    "lfm_audio_quant": LFM_AUDIO_QUANT,
    "lfm_text_id": LFM_TEXT_ID,
    "lfm_text_fallback_id": LFM_TEXT_FALLBACK_ID,
    "bangla_asr_id": BANGLA_ASR_ID,
    "out_root": str(OUT_ROOT),
    "target_sr": TARGET_SR,
    "run_fast": RUN_FAST,
    "random_audit_n": RANDOM_AUDIT_N,
    "clean_demo_n": CLEAN_DEMO_N,
    "stress_n": STRESS_N,
    "lfm_audio_n": LFM_AUDIO_N,
    "lfm_repair_n": LFM_REPAIR_N,
    "action_card_n": ACTION_CARD_N,
    "device": str(DEVICE),
    "dtype": str(DTYPE),
    "gpu_memory": gpu_memory(),
}
safe_json_write(run_manifest, OUT_ROOT / "run_manifest.json")

print("=" * 100)
print("SETUP COMPLETE")
print("=" * 100)
print(json.dumps(run_manifest, indent=2, ensure_ascii=False))

# ============================================================
# 2. Utilities
# ============================================================
BN_DIGITS = "০১২৩৪৫৬৭৮৯"
EN_DIGITS = "0123456789"
EN_TO_BN_DIGIT = str.maketrans(EN_DIGITS, BN_DIGITS)
BN_TO_EN_DIGIT = str.maketrans(BN_DIGITS, EN_DIGITS)

MULTISPACE_RE = re.compile(r"\s+")
PUNCT_RE = re.compile(r"[“”‘’\"'`´।,;:!?()\[\]{}<>/\\|@#$%^&*_+=~]")
BN_CHAR_RE = re.compile(r"[\u0980-\u09FF]")
BAD_UNICODE_RE = re.compile(r"[�৳\u0900-\u097F\u0B00-\u0B7F]")

def normalize_bangla_text(text, digit_mode="bn"):
    if text is None:
        return ""
    import unicodedata
    text = unicodedata.normalize("NFKC", str(text))
    if digit_mode == "bn":
        text = text.translate(EN_TO_BN_DIGIT)
    elif digit_mode == "en":
        text = text.translate(BN_TO_EN_DIGIT)
    text = PUNCT_RE.sub(" ", text)
    text = re.sub(r"[-–—]+", " ", text)
    text = MULTISPACE_RE.sub(" ", text).strip()
    return text

def bn_ratio(text):
    text = str(text)
    chars = [c for c in text if not c.isspace()]
    if not chars:
        return 0.0
    return sum(1 for c in chars if BN_CHAR_RE.match(c)) / max(1, len(chars))

def english_ratio(text):
    text = str(text)
    chars = [c for c in text if not c.isspace()]
    if not chars:
        return 0.0
    return sum(1 for c in chars if ("a" <= c.lower() <= "z")) / max(1, len(chars))

def bangla_quality_ok(text, min_bn_ratio=0.35):
    text = normalize_bangla_text(text)
    if not text:
        return False
    if BAD_UNICODE_RE.search(text):
        return False
    return bn_ratio(text) >= min_bn_ratio

def edit_distance_seq(a, b):
    a, b = list(a), list(b)
    n, m = len(a), len(b)
    if n == 0:
        return m
    if m == 0:
        return n
    prev = list(range(m + 1))
    for i in range(1, n + 1):
        cur = [i] + [0] * m
        ai = a[i - 1]
        for j in range(1, m + 1):
            cost = 0 if ai == b[j - 1] else 1
            cur[j] = min(prev[j] + 1, cur[j - 1] + 1, prev[j - 1] + cost)
        prev = cur
    return prev[m]

def cer(ref, hyp):
    ref = normalize_bangla_text(ref)
    hyp = normalize_bangla_text(hyp)
    if len(ref) == 0:
        return 0.0 if len(hyp) == 0 else 1.0
    return edit_distance_seq(ref, hyp) / max(1, len(ref))

def clipped_cer(ref, hyp):
    return min(1.0, cer(ref, hyp))

def wer_simple(ref, hyp):
    ref_w = normalize_bangla_text(ref).split()
    hyp_w = normalize_bangla_text(hyp).split()
    if len(ref_w) == 0:
        return 0.0 if len(hyp_w) == 0 else 1.0
    return edit_distance_seq(ref_w, hyp_w) / max(1, len(ref_w))

def clipped_wer(ref, hyp):
    return min(1.0, wer_simple(ref, hyp))

def transcript_scores(ref, hyp, prefix=""):
    c = cer(ref, hyp)
    w = wer_simple(ref, hyp)
    cc = min(1.0, c)
    cw = min(1.0, w)
    return {
        f"{prefix}cer": float(c),
        f"{prefix}cer_clipped": float(cc),
        f"{prefix}wer": float(w),
        f"{prefix}wer_clipped": float(cw),
        f"{prefix}char_accuracy": float(max(0.0, 1.0 - cc)),
        f"{prefix}word_accuracy_clipped": float(max(0.0, 1.0 - cw)),
        f"{prefix}exact_match": bool(normalize_bangla_text(ref) == normalize_bangla_text(hyp)),
        f"{prefix}empty_prediction": bool(len(normalize_bangla_text(hyp)) == 0),
        f"{prefix}ref_char_len": int(len(normalize_bangla_text(ref))),
        f"{prefix}pred_char_len": int(len(normalize_bangla_text(hyp))),
    }

def safe_mean(values):
    clean = []
    for v in values:
        try:
            if pd.notna(v):
                clean.append(float(v))
        except Exception:
            pass
    return float(np.mean(clean)) if clean else np.nan

def quality_band_from_cer(x):
    try:
        x = float(x)
    except Exception:
        return "unknown"
    if x <= 0.10:
        return "excellent"
    if x <= 0.25:
        return "good"
    if x <= 0.45:
        return "usable"
    return "poor"

def extractive_summary_bn(text, max_chars=110):
    text = normalize_bangla_text(text)
    if len(text) <= max_chars:
        return text
    cut = text[:max_chars].rsplit(" ", 1)[0].strip()
    return cut if cut else text[:max_chars].strip()

def top_keywords_bn(text, k=6):
    text = normalize_bangla_text(text)
    words = [w for w in text.split() if len(w) >= 3 and bn_ratio(w) > 0.5]
    stop = set("আমি তুমি তিনি তারা আমরা আপনার কিন্তু এবং জন্য মধ্যে থেকে হলে সেটা এটি এই সেই যেন তবে মানে হচ্ছে ছিল আছে করে করা সকল একজন".split())
    freq = {}
    for w in words:
        if w not in stop:
            freq[w] = freq.get(w, 0) + 1
    return [w for w, _ in sorted(freq.items(), key=lambda x: (-x[1], x[0]))[:k]]

def decode_audio_value(value, target_sr=TARGET_SR):
    y, sr = None, None

    if hasattr(value, "get_all_samples"):
        samples = value.get_all_samples()
        data = getattr(samples, "data", None)
        sr = int(getattr(samples, "sample_rate", target_sr))
        arr = data.detach().cpu().numpy() if torch.is_tensor(data) else np.asarray(data)
        if arr.ndim == 2:
            arr = arr.mean(axis=0) if arr.shape[0] <= 8 else arr.mean(axis=1)
        y = arr.astype(np.float32)

    elif isinstance(value, dict):
        if value.get("array", None) is not None:
            y = np.asarray(value["array"], dtype=np.float32)
            sr = int(value.get("sampling_rate", target_sr))
        elif value.get("bytes", None) is not None:
            y, sr = sf.read(io.BytesIO(value["bytes"]), dtype="float32", always_2d=False)
        elif value.get("path", None) is not None and Path(str(value["path"])).exists():
            y, sr = sf.read(str(value["path"]), dtype="float32", always_2d=False)

    elif isinstance(value, (np.ndarray, list, tuple)):
        y = np.asarray(value, dtype=np.float32)
        sr = target_sr

    elif isinstance(value, str):
        p = Path(value)
        if p.exists():
            y, sr = sf.read(str(p), dtype="float32", always_2d=False)

    if y is None or sr is None:
        raise ValueError(f"Could not decode audio value of type={type(value)}")

    y = np.asarray(y, dtype=np.float32)
    if y.ndim > 1:
        y = y.mean(axis=1)
    y = np.nan_to_num(y, nan=0.0, posinf=0.0, neginf=0.0)
    y = np.clip(y, -1.0, 1.0).astype(np.float32)

    if int(sr) != int(target_sr):
        try:
            import scipy.signal
            g = math.gcd(int(sr), int(target_sr))
            y = scipy.signal.resample_poly(y, int(target_sr) // g, int(sr) // g).astype(np.float32)
            sr = target_sr
        except Exception as e:
            raise RuntimeError(f"Resampling failed from {sr} to {target_sr}: {repr(e)}")

    if len(y) == 0:
        raise ValueError("Decoded audio is empty.")

    return y, int(sr)

def run_cmd_bytes(cmd, cwd=None, timeout=900, allow_fail=False):
    cmd = [str(x) for x in cmd]
    print("Running:", " ".join(cmd))
    proc = subprocess.run(
        cmd,
        cwd=str(cwd) if cwd else None,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        timeout=timeout,
    )
    stdout = proc.stdout.decode("utf-8", errors="replace")
    stderr = proc.stderr.decode("utf-8", errors="replace")
    tail = (stdout + "\n" + stderr)[-5000:]
    if tail.strip():
        print(tail)
    payload = {
        "cmd": cmd,
        "cwd": str(cwd) if cwd else None,
        "returncode": int(proc.returncode),
        "stdout_tail": stdout[-5000:],
        "stderr_tail": stderr[-5000:],
    }
    if proc.returncode != 0 and not allow_fail:
        raise RuntimeError((stderr or stdout)[-5000:])
    return payload

# ============================================================
# 3. Load BanSpeech and build subsets
# ============================================================
print("Loading dataset:", DATASET_ID)
ds = load_dataset(DATASET_ID)

first_split = list(ds.keys())[0]
first_row = ds[first_split][0]
cols = list(first_row.keys())

text_candidates = [c for c in ["transcription", "transcript", "sentence", "text", "normalized_transcript"] if c in cols]
audio_candidates = [c for c in ["audio", "speech", "wav", "file_path", "path", "audio_path"] if c in cols]

if not text_candidates:
    text_candidates = [c for c in cols if "text" in c.lower() or "trans" in c.lower()]
if not audio_candidates:
    audio_candidates = [c for c in cols if any(k in c.lower() for k in ["audio", "speech", "wav", "path", "file"])]

if not text_candidates:
    raise RuntimeError(f"No text column found. Columns={cols}")
if not audio_candidates:
    raise RuntimeError(f"No audio column found. Columns={cols}")

TEXT_COL = text_candidates[0]
AUDIO_COL = None
audio_detect_errors = {}

for cand in audio_candidates:
    try:
        _y0, _sr0 = decode_audio_value(ds[first_split][0][cand], TARGET_SR)
        AUDIO_COL = cand
        break
    except Exception as e:
        audio_detect_errors[cand] = repr(e)

if AUDIO_COL is None:
    print("Audio detection errors:", json.dumps(audio_detect_errors, indent=2, ensure_ascii=False))
    raise RuntimeError("Could not decode any candidate audio column.")

rows = []
for split_name in ds.keys():
    dset = ds[split_name]
    for i in range(len(dset)):
        item = dset[i]
        text = str(item.get(TEXT_COL, "")).strip()
        norm = normalize_bangla_text(text)
        if not norm:
            continue
        rows.append({
            "sample_uid": f"{split_name}_{i:06d}",
            "source_split": str(split_name),
            "row_index": int(i),
            "domain": str(split_name),
            "reference_transcript": text,
            "normalized_reference": norm,
            "reference_char_len": len(norm),
            "bn_ratio": bn_ratio(norm),
            "english_ratio": english_ratio(norm),
            "language": "bn",
            "source": "banspeech",
        })

all_df = pd.DataFrame(rows).reset_index(drop=True)
if len(all_df) == 0:
    raise RuntimeError("No dataset rows were built.")

domains = sorted(all_df["domain"].unique().tolist())
domain_to_id = {d: i for i, d in enumerate(domains)}
id_to_domain = {i: d for d, i in domain_to_id.items()}
all_df["domain_id"] = all_df["domain"].map(domain_to_id).astype(int)

train_df, temp_df = train_test_split(
    all_df,
    test_size=0.20,
    random_state=SEED,
    stratify=all_df["domain_id"],
)
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    random_state=SEED,
    stratify=temp_df["domain_id"],
)

train_df = train_df.reset_index(drop=True)
val_df = val_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
train_df["split"] = "train"
val_df["split"] = "val"
test_df["split"] = "test"
all_split_df = pd.concat([train_df, val_df, test_df], axis=0).reset_index(drop=True)

def stratified_sample(df, n, seed=SEED, min_len=3, max_len=260, exclude_ids=None):
    exclude_ids = set(exclude_ids or [])
    base = df[~df["sample_uid"].isin(exclude_ids)].copy()
    base = base[base["reference_char_len"].between(min_len, max_len)].copy()
    if len(base) < min(50, len(df)):
        base = df[~df["sample_uid"].isin(exclude_ids)].copy()

    if n is None or n <= 0 or len(base) <= n:
        return base.sample(frac=1.0, random_state=seed).reset_index(drop=True)

    pieces = []
    for _, g in base.groupby("domain", group_keys=False):
        k = max(1, int(round(n * len(g) / len(base))))
        pieces.append(g.sample(n=min(k, len(g)), random_state=seed))

    out = pd.concat(pieces).sample(frac=1.0, random_state=seed).head(n).reset_index(drop=True)
    return out

def get_audio_for_manifest_row(row):
    raw = ds[str(row["source_split"])][int(row["row_index"])]
    return decode_audio_value(raw[AUDIO_COL], TARGET_SR)

def ensure_wav_for_row(row):
    out_path = AUDIO_CACHE / f"{row['sample_uid']}.wav"
    if out_path.exists() and out_path.stat().st_size > 100:
        return out_path
    y, sr = get_audio_for_manifest_row(row)
    sf.write(str(out_path), y, sr)
    return out_path

def materialize_subset(df, subset_name, duration_min=None, duration_max=None):
    df = df.copy().reset_index(drop=True)
    audio_paths, durations, cache_status = [], [], []
    print(f"Caching {subset_name}: {len(df)} candidate WAV files...")
    for _, row in tqdm(df.iterrows(), total=len(df), desc=f"WAV cache: {subset_name}"):
        try:
            p = ensure_wav_for_row(row)
            info = sf.info(str(p))
            dur = float(info.frames / max(1, info.samplerate))
            audio_paths.append(str(p))
            durations.append(dur)
            cache_status.append("ok")
        except Exception as e:
            log_failure("wav_cache_one", e, {"sample_uid": row.get("sample_uid"), "subset": subset_name})
            audio_paths.append("")
            durations.append(np.nan)
            cache_status.append("failed")

    df["audio_path"] = audio_paths
    df["audio_duration_sec"] = durations
    df["audio_cache_status"] = cache_status
    df = df[df["audio_cache_status"].eq("ok")].reset_index(drop=True)

    if duration_min is not None:
        df = df[df["audio_duration_sec"] >= float(duration_min)].reset_index(drop=True)
    if duration_max is not None:
        df = df[df["audio_duration_sec"] <= float(duration_max)].reset_index(drop=True)

    df["subset"] = subset_name
    keep_cols = [
        "sample_uid", "subset", "audio_path", "reference_transcript", "normalized_reference",
        "reference_char_len", "bn_ratio", "english_ratio", "language", "domain",
        "source", "audio_duration_sec", "source_split", "row_index"
    ]
    return df[keep_cols].copy()

used_ids = set()

random_candidates = stratified_sample(test_df, RANDOM_AUDIT_N + 25, SEED, min_len=5, max_len=180, exclude_ids=used_ids)
random_audit_df = materialize_subset(random_candidates, "random_audit", duration_min=0.8, duration_max=12.0).head(RANDOM_AUDIT_N)
used_ids.update(random_audit_df["sample_uid"].tolist())

clean_pool = test_df[
    test_df["reference_char_len"].between(15, 90)
    & (test_df["bn_ratio"] >= 0.70)
    & (test_df["english_ratio"] <= 0.20)
].copy()
clean_candidates = stratified_sample(clean_pool, CLEAN_DEMO_N + 30, SEED + 1, min_len=15, max_len=90, exclude_ids=used_ids)
clean_demo_df = materialize_subset(clean_candidates, "clean_demo", duration_min=1.0, duration_max=5.0).head(CLEAN_DEMO_N)
used_ids.update(clean_demo_df["sample_uid"].tolist())

stress_pool = test_df[
    (test_df["reference_char_len"] <= 18)
    | (test_df["reference_char_len"] >= 90)
    | (test_df["english_ratio"] > 0.20)
].copy()
if len(stress_pool) < STRESS_N:
    stress_pool = test_df.copy()
stress_candidates = stratified_sample(stress_pool, STRESS_N + 30, SEED + 2, min_len=1, max_len=260, exclude_ids=used_ids)
stress_df = materialize_subset(stress_candidates, "stress", duration_min=0.4, duration_max=15.0).head(STRESS_N)
used_ids.update(stress_df["sample_uid"].tolist())

sample_manifest = pd.concat([random_audit_df, clean_demo_df, stress_df], ignore_index=True)
sample_manifest = sample_manifest.drop_duplicates("sample_uid").reset_index(drop=True)

train_df.to_csv(DATA_DIR / "train_metadata.csv", index=False)
val_df.to_csv(DATA_DIR / "val_metadata.csv", index=False)
test_df.to_csv(DATA_DIR / "test_metadata.csv", index=False)
all_split_df.to_csv(DATA_DIR / "all_metadata.csv", index=False)
sample_manifest.to_csv(DATA_DIR / "sample_manifest_all.csv", index=False)
random_audit_df.to_csv(DATA_DIR / "sample_manifest_random_audit.csv", index=False)
clean_demo_df.to_csv(DATA_DIR / "sample_manifest_clean_demo.csv", index=False)
stress_df.to_csv(DATA_DIR / "sample_manifest_stress.csv", index=False)

dataset_manifest = {
    "created_at": now_utc(),
    "dataset_id": DATASET_ID,
    "text_col": TEXT_COL,
    "audio_col": AUDIO_COL,
    "num_total_rows": int(len(all_df)),
    "num_train": int(len(train_df)),
    "num_val": int(len(val_df)),
    "num_test": int(len(test_df)),
    "num_random_audit": int(len(random_audit_df)),
    "num_clean_demo": int(len(clean_demo_df)),
    "num_stress": int(len(stress_df)),
    "num_total_cached_samples": int(len(sample_manifest)),
    "domains": domains,
    "domain_to_id": domain_to_id,
    "audio_detect_errors": audio_detect_errors,
}
safe_json_write(dataset_manifest, DATA_DIR / "dataset_manifest.json")

print("=" * 100)
print("DATASET STUDIO READY")
print("=" * 100)
print(json.dumps(dataset_manifest, indent=2, ensure_ascii=False))
show(sample_manifest.head())

# ============================================================
# 4. LFM-Audio GGUF download
# ============================================================
AUDIO_GGUF_DIR = MODEL_DIR / "lfm_audio_gguf"
AUDIO_GGUF_DIR.mkdir(parents=True, exist_ok=True)

required_audio_files = {
    "main": f"LFM2.5-Audio-1.5B-{LFM_AUDIO_QUANT}.gguf",
    "mmproj": f"mmproj-LFM2.5-Audio-1.5B-{LFM_AUDIO_QUANT}.gguf",
    "vocoder": f"vocoder-LFM2.5-Audio-1.5B-{LFM_AUDIO_QUANT}.gguf",
    "tokenizer": f"tokenizer-LFM2.5-Audio-1.5B-{LFM_AUDIO_QUANT}.gguf",
}

audio_files, audio_errors = {}, {}
if RUN_LFM_AUDIO_AUDIT:
    for key, filename in required_audio_files.items():
        try:
            path = hf_hub_download(
                repo_id=LFM_AUDIO_GGUF_REPO,
                filename=filename,
                local_dir=str(AUDIO_GGUF_DIR),
                local_dir_use_symlinks=False,
            )
            audio_files[key] = str(Path(path).resolve())
        except Exception as e:
            audio_errors[key] = repr(e)
            log_failure(f"download_lfm_audio_{key}", e, {"filename": filename})
else:
    audio_errors["skipped"] = "RUN_LFM_AUDIO_AUDIT=0"

audio_ready = all(k in audio_files for k in required_audio_files)

gguf_manifest = {
    "created_at": now_utc(),
    "repo": LFM_AUDIO_GGUF_REPO,
    "quant": LFM_AUDIO_QUANT,
    "required_audio_files": required_audio_files,
    "audio_files": audio_files,
    "audio_ready": audio_ready,
    "audio_errors": audio_errors,
}
safe_json_write(gguf_manifest, MODEL_DIR / "lfm_audio_gguf_manifest.json")

print("=" * 100)
print("LFM-AUDIO GGUF DOWNLOAD COMPLETE")
print("=" * 100)
print(json.dumps(gguf_manifest, indent=2, ensure_ascii=False))

# ============================================================
# 5. Build/detect llama-liquid-audio-cli
# ============================================================
def find_executable(names, extra_dirs=None):
    extra_dirs = [Path(x) for x in (extra_dirs or []) if x is not None]
    for name in names:
        p = shutil.which(name)
        if p:
            return str(Path(p).resolve())
    for d in extra_dirs:
        if not d.exists():
            continue
        for name in names:
            for p in d.rglob(name):
                if p.is_file():
                    try:
                        p.chmod(p.stat().st_mode | 0o111)
                    except Exception:
                        pass
                    return str(p.resolve())
    return None

LLAMA_CPP_DIR = Path(os.environ.get("LLAMA_CPP_DIR", str(THIRD_PARTY_DIR / "llama.cpp")))
runner_names_cli = ["llama-liquid-audio-cli", "liquid-audio-cli", "lfm-audio-cli"]
runner_names_server = ["llama-liquid-audio-server", "liquid-audio-server", "lfm-audio-server"]
search_dirs = [
    LLAMA_CPP_DIR,
    LLAMA_CPP_DIR / "build",
    LLAMA_CPP_DIR / "build" / "bin",
    LLAMA_CPP_DIR / "bin",
    THIRD_PARTY_DIR,
    OUT_ROOT,
]

runner_info = {
    "created_at": now_utc(),
    "runner_status": "unavailable",
    "runner_cli": None,
    "runner_server": None,
    "build_attempted": False,
    "build_mode": None,
    "help_probe": None,
    "errors": {},
    "commands": [],
}

runner_cli = find_executable(runner_names_cli, search_dirs)
runner_server = find_executable(runner_names_server, search_dirs)

if runner_cli:
    runner_info.update({"runner_status": "available", "runner_cli": runner_cli, "runner_server": runner_server})
elif RUN_LFM_AUDIO_AUDIT and BUILD_LFM_AUDIO_RUNNER:
    runner_info["build_attempted"] = True
    runner_info["build_mode"] = f"ggml-org/llama.cpp pull/{LLAMA_CPP_AUDIO_PR}/head"
    try:
        if not LLAMA_CPP_DIR.exists():
            runner_info["commands"].append(run_cmd_bytes(
                ["git", "clone", "--depth", "1", "https://github.com/ggml-org/llama.cpp.git", str(LLAMA_CPP_DIR)],
                timeout=900,
            ))

        runner_info["commands"].append(run_cmd_bytes(
            ["git", "fetch", "origin", f"pull/{LLAMA_CPP_AUDIO_PR}/head"],
            cwd=LLAMA_CPP_DIR,
            timeout=900,
        ))
        runner_info["commands"].append(run_cmd_bytes(
            ["git", "checkout", "-B", "lfm_audio_pr", "FETCH_HEAD"],
            cwd=LLAMA_CPP_DIR,
            timeout=300,
        ))
        runner_info["commands"].append(run_cmd_bytes(
            ["git", "submodule", "update", "--init", "--recursive"],
            cwd=LLAMA_CPP_DIR,
            timeout=900,
            allow_fail=True,
        ))

        build_dir = LLAMA_CPP_DIR / "build"
        if os.environ.get("CLEAN_LFM_AUDIO_BUILD", "1") == "1" and build_dir.exists():
            shutil.rmtree(build_dir, ignore_errors=True)

        cmake_args = [
            "cmake", "-S", ".", "-B", "build",
            "-DCMAKE_BUILD_TYPE=Release",
            "-DLLAMA_BUILD_TESTS=OFF",
            "-DGGML_BUILD_TESTS=OFF",
            "-DBUILD_TESTING=OFF",
            "-DLLAMA_BUILD_EXAMPLES=ON",
            "-DLLAMA_BUILD_SERVER=ON",
            "-DGGML_CUDA=OFF",
        ]
        runner_info["commands"].append(run_cmd_bytes(cmake_args, cwd=LLAMA_CPP_DIR, timeout=900))
        runner_info["commands"].append(run_cmd_bytes(
            ["cmake", "--build", "build", "--config", "Release", "--target", "llama-liquid-audio-cli", "-j", str(BUILD_JOBS)],
            cwd=LLAMA_CPP_DIR,
            timeout=3600,
            allow_fail=False,
        ))

        runner_cli = find_executable(runner_names_cli, search_dirs)
        runner_server = find_executable(runner_names_server, search_dirs)
        if runner_cli:
            runner_info.update({"runner_status": "available", "runner_cli": runner_cli, "runner_server": runner_server})
        else:
            runner_info["runner_status"] = "target_built_but_runner_missing"
            runner_info["errors"]["runner_missing"] = "Build completed but executable was not found."

    except Exception as e:
        runner_info["runner_status"] = "build_failed"
        runner_info["errors"]["build"] = repr(e)
        log_failure("lfm_audio_runner_build", e)
else:
    runner_info["runner_status"] = "build_disabled_or_audio_disabled"

if runner_info.get("runner_cli"):
    try:
        runner_info["help_probe"] = run_cmd_bytes([runner_info["runner_cli"], "--help"], timeout=60, allow_fail=True)
    except Exception as e:
        runner_info["help_probe"] = {"error": repr(e)}

safe_json_write(runner_info, RESULTS_DIR / "lfm_audio_runner_info.json")

print("=" * 100)
print("LFM-AUDIO RUNNER AUDIT COMPLETE")
print("=" * 100)
print(json.dumps({
    "runner_status": runner_info["runner_status"],
    "runner_cli": runner_info["runner_cli"],
    "runner_server": runner_info["runner_server"],
    "build_attempted": runner_info["build_attempted"],
    "build_mode": runner_info["build_mode"],
    "errors": runner_info["errors"],
}, indent=2, ensure_ascii=False))

# ============================================================
# 6. LFM-Audio capability audit
# ============================================================
def first_error_signature(stdout, stderr, outfile_text, max_lines=20):
    lines = []
    joined = "\n".join([str(stderr or ""), str(stdout or ""), str(outfile_text or "")])
    for line in joined.splitlines():
        line = line.strip()
        if not line:
            continue
        low = line.lower()
        if any(k in low for k in ["error", "failed", "invalid", "exception", "abort", "usage", "assert"]):
            lines.append(line)
        if len(lines) >= max_lines:
            break
    if not lines:
        for line in joined.splitlines():
            line = line.strip()
            if line:
                lines.append(line)
            if len(lines) >= max_lines:
                break
    return "\n".join(lines[:max_lines])[:3000]

def extract_cli_transcript(stdout, stderr="", outfile_text=""):
    text = "\n".join([str(outfile_text or ""), str(stdout or "")]).strip()
    if not text:
        return ""
    lines = [l.strip() for l in text.splitlines() if l.strip()]
    bad_prefixes = (
        "llama_", "ggml_", "main:", "system_info", "sampling", "build:",
        "warning", "error", "usage:", "options:", "load_", "print_",
        "clip_", "mtmd_", "encode_", "decode_", "log_", "init:", "llm_",
        "model ", "tokenizer "
    )
    candidates = []
    for line in lines:
        low = line.lower()
        if any(low.startswith(p) for p in bad_prefixes):
            continue
        clean = normalize_bangla_text(line)
        if len(clean) >= 2 and bn_ratio(clean) >= 0.20:
            candidates.append(clean)
    return candidates[-1] if candidates else ""

def classify_lfm_audio_failure(returncode, hyp, stdout, stderr, outfile_text):
    joined = "\n".join([str(stdout or ""), str(stderr or ""), str(outfile_text or "")]).lower()
    if hyp:
        return "ok"
    if returncode is None:
        return "subprocess_exception"
    if int(returncode) != 0:
        if "invalid argument" in joined:
            return "cli_argument_error"
        if "usage:" in joined:
            return "cli_usage_error"
        if "error" in joined or "failed" in joined:
            return "cli_runtime_error"
        return "nonzero_returncode"
    if "print_info" in joined or "model params" in joined:
        return "runner_loaded_but_no_transcript"
    if len(joined.strip()) == 0:
        return "empty_stdout_stderr"
    return "no_parseable_transcript"

def build_lfm_audio_cmds(audio_path, out_txt, system_prompt, user_prompt):
    cli = runner_info.get("runner_cli")
    main = audio_files["main"]
    mmproj = audio_files["mmproj"]
    vocoder = audio_files["vocoder"]
    tokenizer_gguf = audio_files["tokenizer"]

    base = [
        cli, "-m", main, "--mmproj", mmproj,
        "-mv", vocoder, "--tts-speaker-file", tokenizer_gguf,
        "-sys", system_prompt, "--audio", audio_path, "-p", user_prompt,
    ]
    return [base, base + ["-o", out_txt]]

def run_lfm_audio_one(row):
    if not RUN_LFM_AUDIO_AUDIT:
        return {"transcript": "", "status": "skipped", "failure_type": "skipped", "runtime_sec": np.nan}
    if runner_info.get("runner_status") != "available":
        return {"transcript": "", "status": "failed", "failure_type": "runner_unavailable", "runtime_sec": np.nan}
    if not audio_ready:
        return {"transcript": "", "status": "failed", "failure_type": "gguf_unavailable", "runtime_sec": np.nan}

    wav = str(row["audio_path"])
    sample_uid = str(row["sample_uid"])
    prompt_pairs = [
        (
            "You are a speech recognition system. Transcribe the provided Bengali audio exactly. Return only Bengali text. Do not translate. Do not explain.",
            "Transcribe this audio in Bengali only."
        ),
        (
            "You are an audio understanding assistant. Listen to the audio and write only the spoken Bengali words.",
            "What Bengali speech is in this audio? Return only the transcript."
        ),
    ]

    last = None
    for pi, (sys_prompt, user_prompt) in enumerate(prompt_pairs):
        out_txt = OUTFILE_DIR / f"{sample_uid}_try{pi}.txt"
        cmds = build_lfm_audio_cmds(wav, str(out_txt), sys_prompt, user_prompt)

        for ci, cmd in enumerate(cmds):
            try:
                t0 = time.time()
                proc = subprocess.run(
                    [str(x) for x in cmd],
                    stdout=subprocess.PIPE,
                    stderr=subprocess.PIPE,
                    timeout=LFM_AUDIO_TIMEOUT,
                )
                runtime = time.time() - t0
                stdout = proc.stdout.decode("utf-8", errors="replace")
                stderr = proc.stderr.decode("utf-8", errors="replace")
                outfile_text = ""
                if out_txt.exists():
                    try:
                        outfile_text = out_txt.read_text(encoding="utf-8", errors="replace")
                    except Exception:
                        outfile_text = ""

                (STDOUT_DIR / f"{sample_uid}_try{pi}_cmd{ci}.stdout.txt").write_text(stdout, encoding="utf-8", errors="replace")
                (STDERR_DIR / f"{sample_uid}_try{pi}_cmd{ci}.stderr.txt").write_text(stderr, encoding="utf-8", errors="replace")

                hyp = extract_cli_transcript(stdout, stderr, outfile_text)
                failure_type = classify_lfm_audio_failure(proc.returncode, hyp, stdout, stderr, outfile_text)
                status = "ok" if hyp else "failed"

                last = {
                    "transcript": hyp,
                    "status": status,
                    "failure_type": failure_type,
                    "runtime_sec": float(runtime) if status == "ok" else np.nan,
                    "returncode": int(proc.returncode),
                    "error_signature": first_error_signature(stdout, stderr, outfile_text),
                    "stderr_tail": stderr[-2500:],
                    "stdout_tail": stdout[-2500:],
                    "outfile_tail": outfile_text[-2500:],
                    "outfile_path": str(out_txt),
                    "cmd": " ".join(map(str, cmd)),
                }
                if status == "ok":
                    return last
            except Exception as e:
                last = {
                    "transcript": "",
                    "status": "failed",
                    "failure_type": "subprocess_exception",
                    "runtime_sec": np.nan,
                    "returncode": None,
                    "error_signature": repr(e),
                    "stderr_tail": repr(e),
                    "stdout_tail": "",
                    "outfile_tail": "",
                    "outfile_path": str(out_txt),
                    "cmd": " ".join(map(str, cmd)),
                }

    return last or {
        "transcript": "",
        "status": "failed",
        "failure_type": "no_candidate_command",
        "runtime_sec": np.nan,
        "returncode": None,
        "error_signature": "",
        "stderr_tail": "",
        "stdout_tail": "",
        "outfile_tail": "",
        "outfile_path": "",
        "cmd": "",
    }

lfm_audio_rows = []
lfm_audio_source = clean_demo_df.head(min(LFM_AUDIO_N, len(clean_demo_df))).copy()
if len(lfm_audio_source) == 0:
    lfm_audio_source = sample_manifest.head(min(LFM_AUDIO_N, len(sample_manifest))).copy()

for _, row in tqdm(lfm_audio_source.iterrows(), total=len(lfm_audio_source), desc="LFM-Audio capability audit"):
    result = run_lfm_audio_one(row)
    ref = row["normalized_reference"]
    hyp = normalize_bangla_text(result.get("transcript", ""))

    rec = {
        "sample_uid": row["sample_uid"],
        "subset": row.get("subset", "lfm_audio_audit"),
        "audio_path": row["audio_path"],
        "language": row["language"],
        "domain": row["domain"],
        "reference": ref,
        "audio_duration_sec": float(row["audio_duration_sec"]),
        "lfm_audio_transcript": hyp,
        "lfm_audio_status": result.get("status", "unknown"),
        "lfm_audio_failure_type": result.get("failure_type", "unknown"),
        "lfm_audio_runtime_sec": result.get("runtime_sec", np.nan),
        "returncode": result.get("returncode", None),
        "error_signature": result.get("error_signature", ""),
        "cmd": result.get("cmd", ""),
        "stderr_tail": result.get("stderr_tail", ""),
        "stdout_tail": result.get("stdout_tail", ""),
        "outfile_tail": result.get("outfile_tail", ""),
        "outfile_path": result.get("outfile_path", ""),
    }
    rec.update(transcript_scores(ref, hyp, prefix="lfm_audio_"))
    lfm_audio_rows.append(rec)

lfm_audio_df = pd.DataFrame(lfm_audio_rows)
lfm_audio_df.to_csv(RESULTS_DIR / "lfm_audio_capability_audit.csv", index=False)
lfm_audio_df.to_csv(RESULTS_DIR / "lfm_audio_predictions.csv", index=False)

lfm_ok = lfm_audio_df[lfm_audio_df["lfm_audio_status"].eq("ok")] if len(lfm_audio_df) else pd.DataFrame()
lfm_audio_summary = {
    "model": LFM_AUDIO_GGUF_REPO,
    "runner_status": runner_info.get("runner_status"),
    "audit_type": "lfm_audio_runner_capability_audit_not_benchmark",
    "n_requested": int(LFM_AUDIO_N),
    "n_rows": int(len(lfm_audio_df)),
    "n_ok": int(len(lfm_ok)),
    "ok_rate": float(len(lfm_ok) / max(1, len(lfm_audio_df))),
    "cer_mean": safe_mean(lfm_ok["lfm_audio_cer"]) if len(lfm_ok) else np.nan,
    "wer_mean": safe_mean(lfm_ok["lfm_audio_wer"]) if len(lfm_ok) else np.nan,
    "failure_type_counts": lfm_audio_df["lfm_audio_failure_type"].value_counts().to_dict() if len(lfm_audio_df) else {},
    "status_counts": lfm_audio_df["lfm_audio_status"].value_counts().to_dict() if len(lfm_audio_df) else {},
    "first_error_signature": str(lfm_audio_df["error_signature"].dropna().iloc[0]) if len(lfm_audio_df) and lfm_audio_df["error_signature"].notna().any() else "",
    "claim_status": "lfm_audio_asr_claim_allowed" if len(lfm_ok) > 0 else "blocked_no_lfm_audio_asr_claim",
    "note": "This is a capability audit. If no transcript is produced, no LFM-Audio ASR quality claim is made.",
}
safe_json_write(lfm_audio_summary, RESULTS_DIR / "lfm_audio_summary.json")
pd.DataFrame([lfm_audio_summary]).to_csv(RESULTS_DIR / "lfm_audio_summary.csv", index=False)

print("=" * 100)
print("LFM-AUDIO CAPABILITY AUDIT COMPLETE")
print("=" * 100)
print(json.dumps(lfm_audio_summary, indent=2, ensure_ascii=False))
show(lfm_audio_df.head(10))

# ============================================================
# 7. BanglaASR baseline
# ============================================================
def run_bangla_asr_batch(source_df):
    rows = []
    if not RUN_BANGLA_ASR:
        return pd.DataFrame(rows)

    print("Loading BanglaASR:", BANGLA_ASR_ID)
    processor = AutoProcessor.from_pretrained(BANGLA_ASR_ID)
    model = AutoModelForSpeechSeq2Seq.from_pretrained(
        BANGLA_ASR_ID,
        torch_dtype=DTYPE,
        low_cpu_mem_usage=True,
    )
    model.to(DEVICE)
    model.eval()

    forced_decoder_ids = None
    try:
        if hasattr(processor, "get_decoder_prompt_ids"):
            forced_decoder_ids = processor.get_decoder_prompt_ids(language="bengali", task="transcribe")
    except Exception:
        forced_decoder_ids = None

    for _, row in tqdm(source_df.iterrows(), total=len(source_df), desc="BanglaASR"):
        try:
            y, sr = sf.read(str(row["audio_path"]), dtype="float32", always_2d=False)
            if y.ndim > 1:
                y = y.mean(axis=1)

            inputs = processor(y, sampling_rate=TARGET_SR, return_tensors="pt", return_attention_mask=True)
            inputs = {k: v.to(DEVICE) for k, v in inputs.items()}
            for k in list(inputs.keys()):
                if torch.is_tensor(inputs[k]) and torch.is_floating_point(inputs[k]):
                    inputs[k] = inputs[k].to(dtype=DTYPE)

            gen_kwargs = {"max_length": 448}
            if forced_decoder_ids is not None:
                gen_kwargs["forced_decoder_ids"] = forced_decoder_ids

            with torch.no_grad():
                pred_ids = model.generate(**inputs, **gen_kwargs)

            hyp = processor.batch_decode(pred_ids, skip_special_tokens=True)[0]
            hyp = normalize_bangla_text(hyp)
            status = "ok" if hyp else "empty"
            err = ""
        except Exception as e:
            hyp = ""
            status = "failed"
            err = repr(e)
            log_failure("bangla_asr_one", e, {"sample_uid": row.get("sample_uid"), "subset": row.get("subset")})

        ref = row["normalized_reference"]
        rec = {
            "sample_uid": row["sample_uid"],
            "subset": row.get("subset", "unknown"),
            "audio_path": row["audio_path"],
            "domain": row["domain"],
            "language": row["language"],
            "reference": ref,
            "reference_char_len": int(row["reference_char_len"]),
            "audio_duration_sec": float(row["audio_duration_sec"]),
            "banglaasr_transcript": hyp,
            "banglaasr_status": status,
            "banglaasr_error": err,
        }
        rec.update(transcript_scores(ref, hyp, prefix="banglaasr_"))
        rec["asr_quality_band"] = quality_band_from_cer(rec["banglaasr_cer"])
        rows.append(rec)

    del model
    del processor
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    return pd.DataFrame(rows)

asr_df = run_bangla_asr_batch(sample_manifest)
asr_df.to_csv(RESULTS_DIR / "banglaasr_predictions_all.csv", index=False)

for subset_name, sdf in asr_df.groupby("subset"):
    sdf.to_csv(RESULTS_DIR / f"banglaasr_predictions_{subset_name}.csv", index=False)

def summarize_asr(df, label):
    ok = df[df["banglaasr_status"].eq("ok")] if len(df) else pd.DataFrame()
    return {
        "subset": label,
        "model": BANGLA_ASR_ID,
        "n_rows": int(len(df)),
        "n_ok": int(len(ok)),
        "ok_rate": float(len(ok) / max(1, len(df))),
        "cer_mean": safe_mean(ok["banglaasr_cer"]) if len(ok) else np.nan,
        "cer_clipped_mean": safe_mean(ok["banglaasr_cer_clipped"]) if len(ok) else np.nan,
        "wer_mean": safe_mean(ok["banglaasr_wer"]) if len(ok) else np.nan,
        "wer_clipped_mean": safe_mean(ok["banglaasr_wer_clipped"]) if len(ok) else np.nan,
        "char_accuracy_mean": safe_mean(ok["banglaasr_char_accuracy"]) if len(ok) else np.nan,
        "exact_match_rate": float(ok["banglaasr_exact_match"].mean()) if len(ok) else np.nan,
        "empty_prediction_rate": float(df["banglaasr_empty_prediction"].mean()) if len(df) else np.nan,
        "quality_band_counts": ok["asr_quality_band"].value_counts().to_dict() if len(ok) else {},
        "claim_status": "baseline speech recognition",
    }

asr_summaries = []
for subset_name, sdf in asr_df.groupby("subset"):
    asr_summaries.append(summarize_asr(sdf, subset_name))
asr_summaries.append(summarize_asr(asr_df, "all"))

asr_summary_df = pd.DataFrame(asr_summaries)
asr_summary_df.to_csv(RESULTS_DIR / "banglaasr_summary_by_subset.csv", index=False)
banglaasr_summary = asr_summary_df[asr_summary_df["subset"].eq("all")].iloc[0].to_dict() if len(asr_summary_df) else {}
safe_json_write({"summaries": asr_summaries}, RESULTS_DIR / "banglaasr_summary.json")

print("=" * 100)
print("BANGLAASR BASELINE COMPLETE")
print("=" * 100)
show(asr_summary_df)
show(asr_df.head(10))

# ============================================================
# 8. Liquid text model
# ============================================================
loaded_lfm_text_id = None
lfm_tokenizer, lfm_model = None, None
lfm_load_error = None

def load_lfm_text_model():
    global loaded_lfm_text_id, lfm_tokenizer, lfm_model, lfm_load_error
    if not RUN_LFM_TEXT:
        lfm_load_error = "RUN_LFM_TEXT=0"
        return None, None

    for model_id in [LFM_TEXT_ID, LFM_TEXT_FALLBACK_ID]:
        try:
            print("Loading Liquid text model:", model_id)
            tok = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
            if tok.pad_token is None:
                tok.pad_token = tok.eos_token

            model = AutoModelForCausalLM.from_pretrained(
                model_id,
                torch_dtype=DTYPE,
                low_cpu_mem_usage=True,
                trust_remote_code=True,
            )
            model.to(DEVICE)
            model.eval()

            loaded_lfm_text_id = model_id
            lfm_tokenizer = tok
            lfm_model = model
            return tok, model
        except Exception as e:
            lfm_load_error = repr(e)
            log_failure("lfm_text_load", e, {"model_id": model_id})
            try:
                del tok
                del model
            except Exception:
                pass
            gc.collect()
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
    return None, None

def lfm_generate(prompt, max_new_tokens=96):
    if lfm_tokenizer is None or lfm_model is None:
        return {"text": "", "status": "model_unavailable", "error": lfm_load_error or ""}

    try:
        if hasattr(lfm_tokenizer, "apply_chat_template") and lfm_tokenizer.chat_template:
            messages = [{"role": "user", "content": prompt}]
            text = lfm_tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
        else:
            text = prompt

        inputs = lfm_tokenizer(text, return_tensors="pt", truncation=True, max_length=1536)
        inputs = {k: v.to(DEVICE) for k, v in inputs.items()}

        with torch.no_grad():
            out_ids = lfm_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                temperature=None,
                top_p=None,
                pad_token_id=lfm_tokenizer.pad_token_id,
                eos_token_id=lfm_tokenizer.eos_token_id,
            )

        gen = out_ids[0][inputs["input_ids"].shape[1]:]
        out = lfm_tokenizer.decode(gen, skip_special_tokens=True).strip()
        return {"text": out, "status": "ok", "error": ""}
    except Exception as e:
        return {"text": "", "status": "failed", "error": repr(e)}

lfm_tokenizer, lfm_model = load_lfm_text_model()

text_model_manifest = {
    "created_at": now_utc(),
    "requested": LFM_TEXT_ID,
    "fallback": LFM_TEXT_FALLBACK_ID,
    "loaded": loaded_lfm_text_id,
    "status": "available" if lfm_model is not None else "unavailable",
    "error": lfm_load_error,
    "gpu_memory": gpu_memory(),
}
safe_json_write(text_model_manifest, MODEL_DIR / "lfm_text_model_manifest.json")

print("=" * 100)
print("LIQUID TEXT MODEL LOAD COMPLETE")
print("=" * 100)
print(json.dumps(text_model_manifest, indent=2, ensure_ascii=False))

# ============================================================
# 9. Liquid repair potential
# ============================================================
def clean_lfm_correction(text):
    text = str(text).strip()
    text = re.sub(r"^```(?:text|bn|bengali)?", "", text, flags=re.I).strip()
    text = re.sub(r"```$", "", text).strip()
    lines = [normalize_bangla_text(x) for x in text.splitlines() if normalize_bangla_text(x)]
    if not lines:
        return ""
    lines = sorted(lines, key=lambda x: (bn_ratio(x), len(x)), reverse=True)
    return normalize_bangla_text(lines[0])

def correction_gate(raw, candidate):
    raw_n = normalize_bangla_text(raw)
    cand_n = normalize_bangla_text(candidate)

    if not cand_n:
        return False, "empty_candidate"
    if not bangla_quality_ok(cand_n):
        return False, "bad_bangla_quality"
    if not raw_n:
        return False, "empty_raw"

    ratio = len(cand_n) / max(1, len(raw_n))
    if ratio < 0.80 or ratio > 1.20:
        return False, f"length_ratio_outside_conservative_range:{ratio:.3f}"

    dist = cer(raw_n, cand_n)
    if dist > 0.12:
        return False, f"too_much_edit_conservative:{dist:.3f}"

    return True, "accepted"

repair_source = asr_df[asr_df["banglaasr_status"].eq("ok")].copy()
subset_priority = {"clean_demo": 0, "random_audit": 1, "stress": 2}
repair_source["subset_priority"] = repair_source["subset"].map(subset_priority).fillna(9)
repair_source = repair_source.sort_values(["subset_priority", "banglaasr_cer"]).head(min(LFM_REPAIR_N, len(repair_source))).copy()

repair_rows = []

for _, row in tqdm(repair_source.iterrows(), total=len(repair_source), desc="Liquid repair potential"):
    raw = normalize_bangla_text(row["banglaasr_transcript"])
    ref = normalize_bangla_text(row["reference"])

    prompt = "\n".join([
        "Correct only obvious Bengali ASR spelling/noise errors.",
        "Return only the corrected Bengali transcript.",
        "Do not translate.",
        "Do not add new facts.",
        "Do not explain.",
        "",
        "Raw Bengali ASR transcript:",
        raw,
    ]).strip()

    out = lfm_generate(prompt, max_new_tokens=96)
    candidate = clean_lfm_correction(out.get("text", ""))

    accepted, gate_reason = correction_gate(raw, candidate)
    deploy_safe = candidate if accepted else raw

    raw_cer = cer(ref, raw)
    candidate_cer = cer(ref, candidate)
    deploy_safe_cer = cer(ref, deploy_safe)

    eval_oracle_best = candidate if candidate and candidate_cer < raw_cer else raw
    eval_oracle_best_cer = min(candidate_cer if candidate else 1.0, raw_cer)

    rec = {
        "sample_uid": row["sample_uid"],
        "subset": row["subset"],
        "domain": row["domain"],
        "reference": ref,
        "raw_asr": raw,
        "lfm_candidate": candidate,
        "deploy_safe_transcript": deploy_safe,
        "eval_oracle_best_transcript": eval_oracle_best,
        "lfm_generation_status": out.get("status", ""),
        "lfm_generation_error": out.get("error", ""),
        "accepted_by_deploy_gate": bool(accepted),
        "gate_reason": gate_reason,
    }
    rec.update(transcript_scores(ref, raw, prefix="raw_"))
    rec.update(transcript_scores(ref, candidate, prefix="candidate_"))
    rec.update(transcript_scores(ref, deploy_safe, prefix="deploy_safe_"))
    rec.update(transcript_scores(ref, eval_oracle_best, prefix="eval_oracle_"))

    rec["candidate_improved"] = bool(candidate and candidate_cer < raw_cer)
    rec["deploy_safe_improved"] = bool(deploy_safe_cer < raw_cer)
    rec["eval_oracle_improved"] = bool(eval_oracle_best_cer < raw_cer)
    rec["candidate_delta_positive_is_better"] = float(raw_cer - candidate_cer) if candidate else np.nan
    rec["deploy_safe_delta_positive_is_better"] = float(raw_cer - deploy_safe_cer)
    rec["eval_oracle_delta_positive_is_better"] = float(raw_cer - eval_oracle_best_cer)
    repair_rows.append(rec)

repair_df = pd.DataFrame(repair_rows)

if len(repair_df):
    deploy_delta = safe_mean(repair_df["deploy_safe_delta_positive_is_better"])
    if pd.notna(deploy_delta) and deploy_delta < 0:
        repair_df["deploy_safe_transcript"] = repair_df["raw_asr"]
        repair_df["accepted_by_deploy_gate"] = False
        repair_df["gate_reason"] = "deploy_safe_disabled_after_negative_eval_delta"
        for idx, row in repair_df.iterrows():
            scores = transcript_scores(row["reference"], row["deploy_safe_transcript"], prefix="deploy_safe_")
            for k, v in scores.items():
                repair_df.loc[idx, k] = v
            repair_df.loc[idx, "deploy_safe_improved"] = False
            repair_df.loc[idx, "deploy_safe_delta_positive_is_better"] = 0.0

repair_df.to_csv(RESULTS_DIR / "lfm_repair_potential_predictions.csv", index=False)
repair_df.to_csv(RESULTS_DIR / "lfm_repair_predictions.csv", index=False)

def summarize_repair(df, label):
    return {
        "subset": label,
        "model": loaded_lfm_text_id or "unavailable",
        "n_rows": int(len(df)),
        "raw_cer_mean": safe_mean(df["raw_cer"]) if len(df) else np.nan,
        "raw_cer_clipped_mean": safe_mean(df["raw_cer_clipped"]) if len(df) else np.nan,
        "candidate_cer_mean": safe_mean(df["candidate_cer"]) if len(df) else np.nan,
        "candidate_cer_clipped_mean": safe_mean(df["candidate_cer_clipped"]) if len(df) else np.nan,
        "deploy_safe_cer_mean": safe_mean(df["deploy_safe_cer"]) if len(df) else np.nan,
        "eval_oracle_cer_mean": safe_mean(df["eval_oracle_cer"]) if len(df) else np.nan,
        "candidate_delta_positive_is_better": safe_mean(df["candidate_delta_positive_is_better"]) if len(df) else np.nan,
        "deploy_safe_delta_positive_is_better": safe_mean(df["deploy_safe_delta_positive_is_better"]) if len(df) else np.nan,
        "eval_oracle_delta_positive_is_better": safe_mean(df["eval_oracle_delta_positive_is_better"]) if len(df) else np.nan,
        "candidate_improvement_rate": float(df["candidate_improved"].mean()) if len(df) else np.nan,
        "deploy_safe_improvement_rate": float(df["deploy_safe_improved"].mean()) if len(df) else np.nan,
        "eval_oracle_improvement_rate": float(df["eval_oracle_improved"].mean()) if len(df) else np.nan,
        "deploy_gate_accept_rate": float(df["accepted_by_deploy_gate"].mean()) if len(df) else np.nan,
        "claim_status": "repair-potential analysis; deploy-safe transcript never claims unsafe automatic improvement",
    }

repair_summaries = []
for subset_name, sdf in (repair_df.groupby("subset") if len(repair_df) else []):
    repair_summaries.append(summarize_repair(sdf, subset_name))
repair_summaries.append(summarize_repair(repair_df, "all"))

repair_summary_df = pd.DataFrame(repair_summaries)
repair_summary_df.to_csv(RESULTS_DIR / "lfm_repair_summary_by_subset.csv", index=False)
repair_summary = repair_summary_df[repair_summary_df["subset"].eq("all")].iloc[0].to_dict() if len(repair_summary_df) else {}
repair_summary["gate_reason_counts"] = repair_df["gate_reason"].value_counts().to_dict() if len(repair_df) else {}
safe_json_write({"summaries": repair_summaries, "gate_reason_counts": repair_summary.get("gate_reason_counts", {})}, RESULTS_DIR / "lfm_repair_summary.json")

print("=" * 100)
print("LIQUID REPAIR-POTENTIAL ANALYSIS COMPLETE")
print("=" * 100)
show(repair_summary_df)
show(repair_df.head(10))

# ============================================================
# 10. Router with abstention
# ============================================================
router_model = Pipeline([
    ("features", FeatureUnion([
        ("charwb", TfidfVectorizer(
            analyzer="char_wb",
            ngram_range=(2, 5),
            min_df=2,
            max_features=140000,
            sublinear_tf=True,
        )),
        ("word", TfidfVectorizer(
            analyzer="word",
            ngram_range=(1, 2),
            min_df=2,
            max_features=50000,
            sublinear_tf=True,
        )),
    ])),
    ("clf", LogisticRegression(
        max_iter=1200,
        class_weight="balanced",
        C=3.0,
        n_jobs=2,
        random_state=SEED,
    )),
])

router_model.fit(train_df["normalized_reference"].astype(str).tolist(), train_df["domain"].astype(str).tolist())

test_probs = router_model.predict_proba(test_df["normalized_reference"].astype(str).tolist())
test_classes = list(router_model.named_steps["clf"].classes_)
test_pred_ids = test_probs.argmax(axis=1)
test_pred = [test_classes[int(i)] for i in test_pred_ids]

router_test_acc = accuracy_score(test_df["domain"].astype(str).tolist(), test_pred)
router_test_f1 = f1_score(test_df["domain"].astype(str).tolist(), test_pred, average="macro")

router_ref_eval = pd.DataFrame({
    "sample_uid": test_df["sample_uid"],
    "reference": test_df["normalized_reference"],
    "true_domain": test_df["domain"],
    "pred_domain": test_pred,
    "confidence": test_probs.max(axis=1),
})
router_ref_eval.to_csv(RESULTS_DIR / "router_reference_eval.csv", index=False)

def router_predict(texts):
    if not texts:
        return [], []
    probs = router_model.predict_proba(texts)
    pred_ids = probs.argmax(axis=1)
    conf = probs.max(axis=1)
    classes = list(router_model.named_steps["clf"].classes_)
    pred_domains = [classes[int(i)] for i in pred_ids]
    return pred_domains, conf

def apply_route_abstention(pred_domain, conf, transcript, asr_cer_value=None):
    reasons = []
    transcript = normalize_bangla_text(transcript)

    if conf < ROUTE_CONF_THRESHOLD:
        reasons.append("low_route_confidence")
    if len(transcript) < MIN_ROUTE_CHARS:
        reasons.append("short_transcript")
    try:
        if asr_cer_value is not None and pd.notna(asr_cer_value) and float(asr_cer_value) > MAX_AUTO_CER:
            reasons.append("high_asr_cer_eval_guard")
    except Exception:
        pass

    if reasons:
        return "unknown_review", "abstained", reasons
    return pred_domain, "accepted", []

router_summary = {
    "model": "char-wb+word TF-IDF + LogisticRegression with unknown_review abstention",
    "train_rows": int(len(train_df)),
    "test_rows": int(len(test_df)),
    "reference_test_accuracy": float(router_test_acc),
    "reference_test_macro_f1": float(router_test_f1),
    "route_conf_threshold": ROUTE_CONF_THRESHOLD,
    "auto_route_conf_threshold": AUTO_ROUTE_CONF_THRESHOLD,
    "min_route_chars": MIN_ROUTE_CHARS,
    "max_auto_cer": MAX_AUTO_CER,
    "claim_status": "deterministic evidence router with abstention",
}
safe_json_write(router_summary, RESULTS_DIR / "router_summary.json")
pd.DataFrame([router_summary]).to_csv(RESULTS_DIR / "router_summary.csv", index=False)

print("=" * 100)
print("EVIDENCE ROUTER COMPLETE")
print("=" * 100)
print(json.dumps(router_summary, indent=2, ensure_ascii=False))

# ============================================================
# 11. Deterministic-first safe action cards
# ============================================================
UNKNOWN_DOMAIN = "unknown_review"

ACTION_BY_DOMAIN = {
    "audio_books": "summarize_passage",
    "biography": "extract_person_note",
    "celebrity_interview": "summarize_interview",
    "class_lecture": "create_study_note",
    "documentary": "summarize_documentary",
    "drama_series": "summarize_dialogue",
    "kid_cartoon": "child_safe_summary",
    "kid_voice": "child_speech_review",
    "medicine": "medical_review_required",
    "parliament_speech": "civic_speech_summary",
    "political_talkshow": "discussion_summary",
    "sports": "sports_summary",
    "television_news": "news_summary",
    UNKNOWN_DOMAIN: "manual_review_required",
}

TASK_BY_DOMAIN = {
    "audio_books": "audio_book_note",
    "biography": "biography_note",
    "celebrity_interview": "interview_note",
    "class_lecture": "lecture_note",
    "documentary": "documentary_summary",
    "drama_series": "dialogue_note",
    "kid_cartoon": "child_content_note",
    "kid_voice": "child_speech_note",
    "medicine": "medical_note_review",
    "parliament_speech": "civic_speech_note",
    "political_talkshow": "discussion_note",
    "sports": "sports_note",
    "television_news": "news_note",
    UNKNOWN_DOMAIN: "manual_review_note",
}

VALID_TASKS = set(TASK_BY_DOMAIN.values())
VALID_ACTIONS = set(ACTION_BY_DOMAIN.values())
VALID_RISKS = {"low", "medium", "high"}

def deterministic_risk(route_domain, confidence, transcript, route_status, asr_cer_value=None):
    transcript = normalize_bangla_text(transcript)
    if route_status == "abstained":
        return "high" if route_domain == UNKNOWN_DOMAIN else "medium"
    if route_domain in {"medicine"}:
        return "high"
    if route_domain in {"political_talkshow", "parliament_speech", "television_news"}:
        return "medium" if confidence >= 0.60 else "high"
    if route_domain in {"kid_voice", "kid_cartoon"}:
        return "medium"
    try:
        if asr_cer_value is not None and pd.notna(asr_cer_value) and float(asr_cer_value) > MAX_AUTO_CER:
            return "high"
    except Exception:
        pass
    if len(transcript) < MIN_ROUTE_CHARS:
        return "medium"
    return "low" if confidence >= AUTO_ROUTE_CONF_THRESHOLD else "medium"

def deterministic_note(route_domain, risk, route_status):
    if route_status == "abstained" or route_domain == UNKNOWN_DOMAIN:
        return "Manual review is required because the transcript evidence is too weak for reliable routing."
    if risk == "high":
        return f"Review this {route_domain.replace('_', ' ')} transcript before any downstream use."
    if risk == "medium":
        return f"Use this {route_domain.replace('_', ' ')} transcript as a reviewable local note."
    return f"This {route_domain.replace('_', ' ')} transcript appears suitable for a low-risk local summary."

def validate_final_card(card):
    required = [
        "sample_uid", "language", "true_domain", "route_domain", "route_confidence",
        "route_status", "task_type", "local_action", "risk", "summary_bn",
        "keywords_bn", "assistant_note_en", "transcript_bn", "raw_asr",
        "needs_review", "review_reasons", "final_json_valid",
    ]
    if not isinstance(card, dict):
        return False
    for k in required:
        if k not in card:
            return False
    if card["task_type"] not in VALID_TASKS:
        return False
    if card["local_action"] not in VALID_ACTIONS:
        return False
    if card["risk"] not in VALID_RISKS:
        return False
    if card["route_status"] not in {"accepted", "abstained"}:
        return False
    if not isinstance(card["keywords_bn"], list):
        return False
    if not isinstance(card["review_reasons"], list):
        return False
    if not isinstance(card["needs_review"], bool):
        return False
    return True

card_source = repair_df[repair_df["subset"].eq("clean_demo")].copy() if len(repair_df) else pd.DataFrame()
if len(card_source) < ACTION_CARD_N and len(repair_df):
    need = ACTION_CARD_N - len(card_source)
    extra = repair_df[~repair_df["sample_uid"].isin(card_source["sample_uid"])].head(need).copy()
    card_source = pd.concat([card_source, extra], ignore_index=True)
card_source = card_source.head(min(ACTION_CARD_N, len(card_source))).reset_index(drop=True)

card_texts = [normalize_bangla_text(x) for x in card_source["deploy_safe_transcript"].astype(str).tolist()] if len(card_source) else []
forced_domains, forced_conf = router_predict(card_texts)

action_card_rows = []
proposal_rows = []
jsonl_path = RESULTS_DIR / "demo_action_cards_safe.jsonl"

for i, (_, row) in enumerate(tqdm(card_source.iterrows(), total=len(card_source), desc="Schema-valid action cards")):
    transcript = normalize_bangla_text(row["deploy_safe_transcript"])
    raw_asr = normalize_bangla_text(row["raw_asr"])
    true_domain = str(row["domain"])
    forced_domain = forced_domains[i] if i < len(forced_domains) else UNKNOWN_DOMAIN
    conf = float(forced_conf[i]) if i < len(forced_conf) else 0.0

    route_domain, route_status, route_reasons = apply_route_abstention(
        forced_domain, conf, transcript, asr_cer_value=row.get("raw_cer", None)
    )

    risk = deterministic_risk(route_domain, conf, transcript, route_status, asr_cer_value=row.get("raw_cer", None))
    deterministic_task = TASK_BY_DOMAIN.get(route_domain, TASK_BY_DOMAIN[UNKNOWN_DOMAIN])
    deterministic_action = ACTION_BY_DOMAIN.get(route_domain, ACTION_BY_DOMAIN[UNKNOWN_DOMAIN])
    summary_bn = extractive_summary_bn(transcript)
    keywords = top_keywords_bn(transcript, k=6)

    review_reasons = list(route_reasons)
    if risk == "high":
        review_reasons.append("high_risk")
    if row.get("raw_cer", 0) > MAX_AUTO_CER:
        review_reasons.append("poor_asr_quality_eval_guard")
    if len(transcript) < MIN_ROUTE_CHARS:
        review_reasons.append("short_transcript")
    review_reasons = sorted(set(review_reasons))

    needs_review = bool(review_reasons or risk in {"medium", "high"})
    assistant_note = deterministic_note(route_domain, risk, route_status)

    final_card = {
        "sample_uid": str(row["sample_uid"]),
        "language": "bn",
        "subset": str(row["subset"]),
        "true_domain": true_domain,
        "forced_route_domain": forced_domain,
        "route_domain": route_domain,
        "route_confidence": round(conf, 6),
        "route_status": route_status,
        "task_type": deterministic_task,
        "local_action": deterministic_action,
        "risk": risk,
        "summary_bn": summary_bn,
        "keywords_bn": keywords,
        "assistant_note_en": assistant_note,
        "transcript_bn": transcript,
        "raw_asr": raw_asr,
        "lfm_candidate_repair": normalize_bangla_text(row.get("lfm_candidate", "")),
        "asr_quality_band": quality_band_from_cer(row.get("raw_cer", np.nan)),
        "raw_cer_eval": float(row.get("raw_cer", np.nan)) if pd.notna(row.get("raw_cer", np.nan)) else None,
        "needs_review": needs_review,
        "review_reasons": review_reasons,
        "final_json_valid": True,
    }
    final_card["final_json_valid"] = validate_final_card(final_card)
    action_card_rows.append(final_card)

action_cards_df = pd.DataFrame(action_card_rows)
proposal_df = pd.DataFrame(proposal_rows)

with jsonl_path.open("w", encoding="utf-8") as f:
    for card in action_card_rows:
        f.write(json.dumps(to_builtin(card), ensure_ascii=False) + "\n")

action_cards_df.to_csv(RESULTS_DIR / "lfm_action_cards_safe.csv", index=False)
proposal_df.to_csv(RESULTS_DIR / "lfm_card_metadata_proposals.csv", index=False)

if len(action_cards_df):
    forced_route_accuracy = accuracy_score(action_cards_df["true_domain"], action_cards_df["forced_route_domain"])
    accepted = action_cards_df[action_cards_df["route_status"].eq("accepted")]
    coverage = float(len(accepted) / len(action_cards_df))
    abstention_rate = 1.0 - coverage
    accuracy_on_covered = accuracy_score(accepted["true_domain"], accepted["route_domain"]) if len(accepted) else np.nan
    macro_f1_on_covered = f1_score(accepted["true_domain"], accepted["route_domain"], average="macro") if len(accepted) else np.nan
else:
    forced_route_accuracy = np.nan
    coverage = np.nan
    abstention_rate = np.nan
    accuracy_on_covered = np.nan
    macro_f1_on_covered = np.nan

card_summary = {
    "n_rows": int(len(action_cards_df)),
    "forced_route_accuracy_before_abstention": float(forced_route_accuracy) if pd.notna(forced_route_accuracy) else np.nan,
    "coverage_after_abstention": float(coverage) if pd.notna(coverage) else np.nan,
    "abstention_rate": float(abstention_rate) if pd.notna(abstention_rate) else np.nan,
    "accuracy_on_covered": float(accuracy_on_covered) if pd.notna(accuracy_on_covered) else np.nan,
    "macro_f1_on_covered": float(macro_f1_on_covered) if pd.notna(macro_f1_on_covered) else np.nan,
    "final_json_valid_rate": float(action_cards_df["final_json_valid"].mean()) if len(action_cards_df) else 0.0,
    "needs_review_rate": float(action_cards_df["needs_review"].mean()) if len(action_cards_df) else 0.0,
    "risk_counts": action_cards_df["risk"].value_counts().to_dict() if len(action_cards_df) else {},
    "route_status_counts": action_cards_df["route_status"].value_counts().to_dict() if len(action_cards_df) else {},
    "route_domain_counts": action_cards_df["route_domain"].value_counts().to_dict() if len(action_cards_df) else {},
    "claim_status": "deterministic schema-valid reviewable speech-assistant dataset cards",
    "note": "Cards are deterministic-first; Liquid proposals are optional and saved separately.",
}
safe_json_write(card_summary, RESULTS_DIR / "lfm_action_cards_summary.json")
pd.DataFrame([card_summary]).to_csv(RESULTS_DIR / "lfm_action_cards_summary.csv", index=False)

print("=" * 100)
print("SAFE ACTION CARDS COMPLETE")
print("=" * 100)
print(json.dumps(card_summary, indent=2, ensure_ascii=False))
show(action_cards_df.head(10))

# ============================================================
# 12. Figures
# ============================================================
try:
    domain_counts = all_df["domain"].value_counts().reindex(domains).reset_index()
    domain_counts.columns = ["domain", "count"]
    domain_counts.to_csv(RESULTS_DIR / "dataset_domain_distribution.csv", index=False)

    plt.figure(figsize=(10, 4))
    plt.bar(domain_counts["domain"], domain_counts["count"])
    plt.xticks(rotation=75, ha="right")
    plt.ylabel("Rows")
    plt.title("BanSpeech domain distribution")
    plt.tight_layout()
    plt.savefig(FIGURES_DIR / "dataset_domain_distribution.png", dpi=180)
    plt.show()

    if len(asr_df):
        domain_asr = (
            asr_df[asr_df["banglaasr_status"].eq("ok")]
            .groupby(["subset", "domain"])[["banglaasr_cer", "banglaasr_cer_clipped", "banglaasr_wer"]]
            .mean()
            .reset_index()
        )
        domain_asr.to_csv(RESULTS_DIR / "banglaasr_domain_metrics.csv", index=False)

        subset_cer = asr_summary_df[asr_summary_df["subset"].ne("all")].copy()
        plt.figure(figsize=(7, 4))
        plt.bar(subset_cer["subset"], subset_cer["cer_clipped_mean"])
        plt.ylabel("Clipped CER")
        plt.title("BanglaASR clipped CER by subset")
        plt.tight_layout()
        plt.savefig(FIGURES_DIR / "banglaasr_clipped_cer_by_subset.png", dpi=180)
        plt.show()

    if len(repair_summary_df):
        all_repair = repair_summary_df[repair_summary_df["subset"].eq("all")]
        if len(all_repair):
            vals = [
                float(all_repair["raw_cer_clipped_mean"].iloc[0]),
                float(all_repair["candidate_cer_clipped_mean"].iloc[0]),
                min(1.0, float(all_repair["deploy_safe_cer_mean"].iloc[0])),
                min(1.0, float(all_repair["eval_oracle_cer_mean"].iloc[0])),
            ]
            labels = ["Raw ASR", "LFM candidate", "Deploy-safe", "Eval-oracle"]
            plt.figure(figsize=(8, 4))
            plt.bar(labels, vals)
            plt.ylabel("CER / clipped CER")
            plt.title("Liquid transcript repair potential")
            plt.tight_layout()
            plt.savefig(FIGURES_DIR / "repair_potential_bar.png", dpi=180)
            plt.show()

    if len(action_cards_df):
        status_counts = action_cards_df["route_status"].value_counts().reset_index()
        status_counts.columns = ["route_status", "count"]
        plt.figure(figsize=(6, 4))
        plt.bar(status_counts["route_status"], status_counts["count"])
        plt.ylabel("Count")
        plt.title("Router coverage vs abstention")
        plt.tight_layout()
        plt.savefig(FIGURES_DIR / "router_abstention_counts.png", dpi=180)
        plt.show()

        risk_counts = action_cards_df["risk"].value_counts().reset_index()
        risk_counts.columns = ["risk", "count"]
        plt.figure(figsize=(6, 4))
        plt.bar(risk_counts["risk"], risk_counts["count"])
        plt.ylabel("Count")
        plt.title("Action-card risk distribution")
        plt.tight_layout()
        plt.savefig(FIGURES_DIR / "card_risk_distribution.png", dpi=180)
        plt.show()

    if len(lfm_audio_df):
        failure_counts = lfm_audio_df["lfm_audio_failure_type"].value_counts().reset_index()
        failure_counts.columns = ["failure_type", "count"]
        failure_counts.to_csv(RESULTS_DIR / "lfm_audio_failure_type_counts.csv", index=False)

        plt.figure(figsize=(9, 4))
        plt.bar(failure_counts["failure_type"], failure_counts["count"])
        plt.xticks(rotation=45, ha="right")
        plt.ylabel("Count")
        plt.title("LFM-Audio capability audit failure types")
        plt.tight_layout()
        plt.savefig(FIGURES_DIR / "lfm_audio_failure_types.png", dpi=180)
        plt.show()

    if len(sample_manifest):
        y, sr = sf.read(str(sample_manifest.iloc[0]["audio_path"]), dtype="float32", always_2d=False)
        if y.ndim > 1:
            y = y.mean(axis=1)
        plt.figure(figsize=(9, 4))
        plt.specgram(y, Fs=sr, NFFT=512, noverlap=384)
        plt.xlabel("Time (s)")
        plt.ylabel("Frequency (Hz)")
        plt.title("Sample BanSpeech audio spectrogram")
        plt.tight_layout()
        plt.savefig(FIGURES_DIR / "sample_spectrogram.png", dpi=180)
        plt.show()

except Exception as e:
    log_failure("figures", e)

# ============================================================
# 13. Failure casebook + final artifacts
# ============================================================
def md_table(df, max_rows=10):
    if df is None or len(df) == 0:
        return "_No rows._"
    try:
        return df.head(max_rows).to_markdown(index=False)
    except Exception:
        return df.head(max_rows).to_string(index=False)

failure_casebook_lines = [
    "# Failure Casebook",
    "",
    "## 1. LFM-Audio capability audit",
    "",
    "The LFM-Audio section is treated as a capability audit, not an ASR benchmark.",
    "",
    "### LFM-Audio summary",
    "",
    json.dumps(to_builtin(lfm_audio_summary), indent=2, ensure_ascii=False),
    "",
    "## 2. High-CER ASR examples",
    "",
    md_table(asr_df.sort_values("banglaasr_cer", ascending=False)[
        ["subset", "domain", "reference", "banglaasr_transcript", "banglaasr_cer", "banglaasr_cer_clipped", "banglaasr_wer"]
    ], max_rows=12) if len(asr_df) else "_No ASR rows._",
    "",
    "## 3. Liquid repair-potential examples",
    "",
    md_table(repair_df.sort_values("eval_oracle_delta_positive_is_better", ascending=False)[
        ["subset", "domain", "reference", "raw_asr", "lfm_candidate", "raw_cer", "candidate_cer", "eval_oracle_delta_positive_is_better"]
    ], max_rows=12) if len(repair_df) else "_No repair rows._",
    "",
    "## 4. Router abstention examples",
    "",
    md_table(action_cards_df[action_cards_df["route_status"].eq("abstained")][
        ["subset", "true_domain", "forced_route_domain", "route_confidence", "route_domain", "review_reasons", "transcript_bn"]
    ], max_rows=12) if len(action_cards_df) else "_No card rows._",
    "",
    "## 5. High-risk action-card examples",
    "",
    md_table(action_cards_df[action_cards_df["risk"].eq("high")][
        ["subset", "true_domain", "route_domain", "risk", "review_reasons", "assistant_note_en", "transcript_bn"]
    ], max_rows=12) if len(action_cards_df) else "_No card rows._",
]
failure_casebook = "\n".join(failure_casebook_lines)
(RESULTS_DIR / "failure_casebook.md").write_text(failure_casebook, encoding="utf-8")

final_rows = []

final_rows.append({
    "component": "LFM-Audio capability audit",
    "model": LFM_AUDIO_GGUF_REPO,
    "subset": "clean_demo_sample",
    "n": lfm_audio_summary.get("n_rows", 0),
    "primary_metric": "ok_rate",
    "primary_value": lfm_audio_summary.get("ok_rate", np.nan),
    "secondary_metric": "failure_type_counts",
    "secondary_value": json.dumps(lfm_audio_summary.get("failure_type_counts", {}), ensure_ascii=False),
    "claim_status": lfm_audio_summary.get("claim_status", ""),
})

for _, row in asr_summary_df.iterrows():
    final_rows.append({
        "component": "BanglaASR speech front-end",
        "model": BANGLA_ASR_ID,
        "subset": row["subset"],
        "n": row["n_rows"],
        "primary_metric": "clipped_CER",
        "primary_value": row["cer_clipped_mean"],
        "secondary_metric": "raw_CER",
        "secondary_value": row["cer_mean"],
        "claim_status": row["claim_status"],
    })

for _, row in repair_summary_df.iterrows():
    final_rows.append({
        "component": "Liquid repair potential",
        "model": loaded_lfm_text_id or "unavailable",
        "subset": row["subset"],
        "n": row["n_rows"],
        "primary_metric": "eval_oracle_delta",
        "primary_value": row["eval_oracle_delta_positive_is_better"],
        "secondary_metric": "deploy_safe_delta",
        "secondary_value": row["deploy_safe_delta_positive_is_better"],
        "claim_status": row["claim_status"],
    })

final_rows.append({
    "component": "Evidence router",
    "model": "char-wb+word TF-IDF",
    "subset": "reference_test",
    "n": router_summary.get("test_rows", 0),
    "primary_metric": "reference_test_accuracy",
    "primary_value": router_summary.get("reference_test_accuracy", np.nan),
    "secondary_metric": "reference_test_macro_f1",
    "secondary_value": router_summary.get("reference_test_macro_f1", np.nan),
    "claim_status": router_summary.get("claim_status", ""),
})

final_rows.append({
    "component": "Safe action cards",
    "model": "deterministic schema + optional Liquid proposal",
    "subset": "clean_demo_first",
    "n": card_summary.get("n_rows", 0),
    "primary_metric": "final_json_valid_rate",
    "primary_value": card_summary.get("final_json_valid_rate", np.nan),
    "secondary_metric": "accuracy_on_covered",
    "secondary_value": card_summary.get("accuracy_on_covered", np.nan),
    "claim_status": card_summary.get("claim_status", ""),
})

final_comparison = pd.DataFrame(final_rows)
final_comparison.to_csv(RESULTS_DIR / "final_pr_comparison_table.csv", index=False)

error_parts = []
if len(repair_df):
    tmp = repair_df.sort_values("raw_cer", ascending=False).head(30).copy()
    tmp["error_type"] = "high_raw_asr_cer"
    error_parts.append(tmp)

if len(action_cards_df):
    tmp = action_cards_df[action_cards_df["needs_review"]].head(30).copy()
    tmp["error_type"] = "needs_review_action_card"
    error_parts.append(tmp)

if len(lfm_audio_df):
    tmp = lfm_audio_df[~lfm_audio_df["lfm_audio_status"].eq("ok")].head(30).copy()
    tmp["error_type"] = "lfm_audio_failed_or_blocked"
    error_parts.append(tmp)

error_df = pd.concat(error_parts, ignore_index=True, sort=False) if error_parts else pd.DataFrame()
error_df.to_csv(RESULTS_DIR / "error_analysis.csv", index=False)

def df_markdown_safe(df):
    try:
        return df.to_markdown(index=False)
    except Exception:
        return df.to_string(index=False)

final_table_md = df_markdown_safe(final_comparison)

readme_lines = [
    "# LFM Speech Dataset Studio",
    "",
    "A local-first Bangla speech dataset workflow for LFM-Audio capability auditing, baseline ASR analysis, Liquid text repair-potential measurement, abstaining evidence routing, and schema-valid reviewable action cards.",
    "",
    "## Core contribution",
    "",
    "This project does not force unsupported audio-model claims. It audits the LFM-Audio local runner, reports baseline ASR quality honestly, measures Liquid text repair potential separately from deploy-safe correction, and creates reviewable JSONL action cards with deterministic schema hardening.",
    "",
    "## What this project demonstrates",
    "",
    "1. Loads `SUST-CSE-Speech/banspeech`.",
    "2. Creates three evaluation subsets: random audit, clean demo, and stress.",
    "3. Downloads and audits `LiquidAI/LFM2.5-Audio-1.5B-GGUF`.",
    "4. Builds or detects `llama-liquid-audio-cli`.",
    "5. Blocks LFM-Audio ASR claims unless valid transcripts are produced.",
    "6. Runs `bangla-speech-processing/BanglaASR` as a baseline ASR front-end.",
    "7. Uses Liquid text models for transcript repair-potential analysis.",
    "8. Adds `unknown_review` abstention for weak routing evidence.",
    "9. Generates schema-valid reviewable action cards.",
    "",
    "## Key results",
    "",
    final_table_md,
    "",
    "## Output files",
    "",
    "- `final_pr_comparison_table.csv`",
    "- `lfm_audio_capability_audit.csv`",
    "- `banglaasr_predictions_all.csv`",
    "- `banglaasr_summary_by_subset.csv`",
    "- `lfm_repair_potential_predictions.csv`",
    "- `lfm_repair_summary_by_subset.csv`",
    "- `lfm_action_cards_safe.csv`",
    "- `demo_action_cards_safe.jsonl`",
    "- `failure_casebook.md`",
    "- `figures/`",
    "",
    "## Claim discipline",
    "",
    "- LFM-Audio ASR is claimed only if `ok_rate > 0`.",
    "- Failed LFM-Audio runs are retained as capability-audit evidence.",
    "- BanglaASR is reported as a baseline, not a Liquid model.",
    "- Liquid repair is reported as candidate, deploy-safe, and eval-oracle potential.",
    "- Deploy-safe repair must not worsen mean CER.",
    "- Router abstains to `unknown_review` when transcript evidence is weak.",
    "- JSONL action cards are reviewable artifacts, not autonomous decisions.",
    "",
    "## Reproduce on Kaggle",
    "",
    "Settings:",
    "",
    "- Internet: ON",
    "- GPU: T4",
    "- Run the notebook cell once.",
]
readme = "\n".join(readme_lines)

model_card_lines = [
    "# Model Card: LFM Speech Dataset Studio",
    "",
    "## Models",
    "",
    "- Audio audit model: `{}`".format(LFM_AUDIO_GGUF_REPO),
    "- Liquid text model: `{}`".format(loaded_lfm_text_id or "unavailable"),
    "- ASR baseline: `{}`".format(BANGLA_ASR_ID),
    "- Evidence router: char-wb+word TF-IDF + LogisticRegression with abstention",
    "",
    "## Dataset",
    "",
    "- Dataset: `{}`".format(DATASET_ID),
    "- Total rows: {}".format(len(all_df)),
    "- Domains: {}".format(len(domains)),
    "- Random audit rows: {}".format(len(random_audit_df)),
    "- Clean demo rows: {}".format(len(clean_demo_df)),
    "- Stress rows: {}".format(len(stress_df)),
    "",
    "## Metrics",
    "",
    final_table_md,
    "",
    "## Intended use",
    "",
    "Cookbook/community demonstration for local speech-dataset processing, LFM-Audio capability auditing, Liquid text repair-potential analysis, deterministic routing, and schema-valid reviewable action-card generation.",
    "",
    "## Out-of-scope use",
    "",
    "- Do not use action cards as autonomous medical, legal, political, or child-safety decisions.",
    "- Do not claim LFM-Audio Bangla ASR performance unless the audit produces valid transcripts.",
    "- Do not treat BanglaASR metrics as Liquid model metrics.",
    "",
    "## Safety",
    "",
    "The workflow uses explicit claim discipline, byte-safe CLI logging, output-file capture, conservative deploy-safe repair, eval-oracle repair potential reporting, router abstention, deterministic card text, schema-valid JSONL fallback, and `needs_review` flags.",
]
model_card = "\n".join(model_card_lines)

contribution_summary_lines = [
    "# Contribution Summary",
    "",
    "## Proposed title",
    "",
    "LFM Speech Dataset Studio: Bangla Speech Audit, Repair Potential, and Reviewable Action Cards",
    "",
    "## PR type",
    "",
    "Recommended first PR: Community Project entry in the Liquid cookbook linking to the full project repository and artifacts.",
    "",
    "## Why this is useful",
    "",
    "This project demonstrates a realistic local-first low-resource speech workflow around Liquid models while avoiding unsupported benchmark claims.",
    "",
    "## What can be claimed",
    "",
    "- LFM-Audio GGUF files can be downloaded and the local runner can be audited.",
    "- BanglaASR provides a baseline speech front-end.",
    "- Liquid text models can be evaluated for transcript repair potential.",
    "- The router abstains when evidence is weak.",
    "- The final action cards are schema-valid and reviewable.",
    "",
    "## What must not be claimed",
    "",
    "- Do not claim LFM-Audio Bangla ASR accuracy unless `ok_rate > 0`.",
    "- Do not claim deploy-safe Liquid repair improves ASR unless deploy-safe CER delta is positive.",
    "- Do not treat reviewable cards as autonomous decisions.",
]
contribution_summary = "\n".join(contribution_summary_lines)

PR_DIR.mkdir(parents=True, exist_ok=True)
(PR_DIR / "README.md").write_text(readme, encoding="utf-8")
(PR_DIR / "MODEL_CARD.md").write_text(model_card, encoding="utf-8")
(PR_DIR / "CONTRIBUTION_SUMMARY.md").write_text(contribution_summary, encoding="utf-8")
(PR_DIR / "failure_casebook.md").write_text(failure_casebook, encoding="utf-8")
(PR_DIR / "requirements.txt").write_text(
    "\n".join([
        "datasets",
        "transformers",
        "accelerate",
        "huggingface_hub",
        "soundfile",
        "scikit-learn",
        "matplotlib",
        "scipy",
        "tabulate",
    ]),
    encoding="utf-8",
)
(PR_DIR / "pyproject.toml").write_text(
    "\n".join([
        "[project]",
        'name = "lfm-speech-dataset-studio"',
        'version = "0.1.0"',
        'description = "Local-first Bangla speech dataset studio for Liquid model capability auditing and reviewable action cards."',
        'requires-python = ">=3.10"',
    ]),
    encoding="utf-8",
)

key_files = [
    RESULTS_DIR / "final_pr_comparison_table.csv",
    RESULTS_DIR / "lfm_audio_capability_audit.csv",
    RESULTS_DIR / "lfm_audio_summary.csv",
    RESULTS_DIR / "banglaasr_predictions_all.csv",
    RESULTS_DIR / "banglaasr_summary_by_subset.csv",
    RESULTS_DIR / "lfm_repair_potential_predictions.csv",
    RESULTS_DIR / "lfm_repair_summary_by_subset.csv",
    RESULTS_DIR / "lfm_action_cards_safe.csv",
    RESULTS_DIR / "lfm_action_cards_summary.csv",
    RESULTS_DIR / "lfm_card_metadata_proposals.csv",
    RESULTS_DIR / "router_summary.csv",
    RESULTS_DIR / "router_reference_eval.csv",
    RESULTS_DIR / "error_analysis.csv",
    RESULTS_DIR / "failure_casebook.md",
    RESULTS_DIR / "demo_action_cards_safe.jsonl",
    DATA_DIR / "sample_manifest_all.csv",
    DATA_DIR / "sample_manifest_random_audit.csv",
    DATA_DIR / "sample_manifest_clean_demo.csv",
    DATA_DIR / "sample_manifest_stress.csv",
    DATA_DIR / "dataset_manifest.json",
    OUT_ROOT / "run_manifest.json",
]

for src in key_files:
    if Path(src).exists():
        safe_copy(src, PR_DIR / Path(src).name)

# Ensure canonical demo JSONL exists in PR dir.
safe_copy(RESULTS_DIR / "demo_action_cards_safe.jsonl", PR_DIR / "demo_action_cards_safe.jsonl")

fig_dst = PR_DIR / "figures"
fig_dst.mkdir(parents=True, exist_ok=True)
if FIGURES_DIR.exists():
    for p in FIGURES_DIR.glob("*"):
        if p.is_file():
            safe_copy(p, fig_dst / p.name)

tests_dir = PR_DIR / "tests"
tests_dir.mkdir(exist_ok=True)
(tests_dir / "test_card_schema.py").write_text(
    "\n".join([
        "import json",
        "from pathlib import Path",
        "",
        "def test_action_cards_are_valid_jsonl():",
        "    p = Path(__file__).resolve().parents[1] / 'demo_action_cards_safe.jsonl'",
        "    assert p.exists()",
        "    rows = [json.loads(x) for x in p.read_text(encoding='utf-8').splitlines() if x.strip()]",
        "    assert rows",
        "    required = {'sample_uid','language','route_domain','route_status','task_type','local_action','risk','transcript_bn','needs_review','review_reasons','final_json_valid'}",
        "    for r in rows:",
        "        assert required.issubset(r)",
        "        assert r['final_json_valid'] is True",
    ]),
    encoding="utf-8",
)
(tests_dir / "test_claim_discipline.py").write_text(
    "\n".join([
        "from pathlib import Path",
        "",
        "def test_lfm_audio_summary_exists():",
        "    p = Path(__file__).resolve().parents[1] / 'lfm_audio_summary.csv'",
        "    assert p.exists()",
    ]),
    encoding="utf-8",
)

zip_base = OUT_ROOT / "lfm_speech_dataset_studio_v5_1_pr_artifacts"
try:
    shutil.make_archive(str(zip_base), "zip", PR_DIR)
    zip_path = str(zip_base) + ".zip"
except Exception as e:
    log_failure("make_zip", e)
    zip_path = None

final_audit = {
    "created_at": now_utc(),
    "project": "LFM Speech Dataset Studio v5.1",
    "out_root": str(OUT_ROOT),
    "pr_dir": str(PR_DIR),
    "zip_path": zip_path,
    "final_comparison": str(RESULTS_DIR / "final_pr_comparison_table.csv"),
    "readme": str(PR_DIR / "README.md"),
    "model_card": str(PR_DIR / "MODEL_CARD.md"),
    "contribution_summary": str(PR_DIR / "CONTRIBUTION_SUMMARY.md"),
    "demo_action_cards": str(PR_DIR / "demo_action_cards_safe.jsonl"),
    "failure_casebook": str(PR_DIR / "failure_casebook.md"),
    "lfm_audio_summary": lfm_audio_summary,
    "asr_summary": asr_summaries,
    "repair_summary": repair_summary,
    "router_summary": router_summary,
    "card_summary": card_summary,
    "gpu_memory": gpu_memory(),
}
safe_json_write(final_audit, OUT_ROOT / "final_audit.json")
safe_copy(OUT_ROOT / "final_audit.json", PR_DIR / "final_audit.json")

print("=" * 100)
print("NOTEBOOK COMPLETE — LFM SPEECH DATASET STUDIO v5.1")
print("=" * 100)
print(json.dumps({
    "out_root": str(OUT_ROOT),
    "pr_dir": str(PR_DIR),
    "zip_path": zip_path,
    "final_comparison": str(RESULTS_DIR / "final_pr_comparison_table.csv"),
    "readme": str(PR_DIR / "README.md"),
    "model_card": str(PR_DIR / "MODEL_CARD.md"),
    "contribution_summary": str(PR_DIR / "CONTRIBUTION_SUMMARY.md"),
    "demo_action_cards": str(PR_DIR / "demo_action_cards_safe.jsonl"),
    "failure_casebook": str(PR_DIR / "failure_casebook.md"),
    "lfm_audio_claim_status": lfm_audio_summary.get("claim_status", "unknown"),
    "card_json_valid_rate": card_summary.get("final_json_valid_rate", None),
    "router_abstention_rate": card_summary.get("abstention_rate", None),
    "gpu_memory": gpu_memory(),
}, indent=2, ensure_ascii=False))

show(final_comparison)

print("\nREADME preview:\n")
print(readme[:4000])
print("\nPR artifact zip:", zip_path)
````
